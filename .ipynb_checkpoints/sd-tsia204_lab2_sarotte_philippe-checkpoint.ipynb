```python
def mean_covariates(df):
    L=[]
    for i in range(df.shape[1]-1):
        L.append(np.mean(df.iloc[:,i]))
    return L
def standard_deviation_covariates(df):
    L=[]
    for i in range(df.shape[1]-1):
        L.append(np.std(df.iloc[:,i]))
    return L
standard_deviation_covariates(df)
```

## Variable selection 

### 2) 

From now on, we considere that the data has been generated according to a fix model design, more precisely a gaussian model 

```python
def gram_matrix(X,k):### puisque la data est normalise la gram matrix est la matrice constitue de 1 partout
    X=selection_colonne(X,k)
    return (X.T @ X)/np.shape(X)[0]
def unbiased_noise_estimation(X,Y,k,a):
    reg= a
    L=Y-reg.predict(selection_colonne(X,k))
    res_bis=np.linalg.norm(L)**2
    return res_bis/(len(Y)-2)
def OLS_per_covariates(X,Y,k):
    X=selection_colonne(X,k)
    reg= LinearRegression(fit_intercept=False).fit(X,Y)
    return reg
    

def stat_of_no_effect(X,Y,k,a):
    s=np.sqrt(np.linalg.inv(gram_matrix(X,k))[0][0])
    return np.sqrt(X.shape[0])*abs(a.coef_[0])/(s*unbiased_noise_estimation(X,Y,k,a))



def p_value_bis(X,Y,k,a):
    return 2*(1-stats.t.cdf(stat_of_no_effect(X,Y,k,a),len(Y)-np.shape(X)[1]-1))


def selection_colonne(X,k):
    X=np.array(X)
    res =[]
    for i in range(np.shape(X)[0]):
        res.append(X[i][k])
    return np.reshape(res,(-1,1))



def forward_variable_selection(X,Y):
    res=[]
    Y_temp=np.array(Y)
    deja_utilise=[]
    for i in range(np.shape(X)[1]):
        temp_valeur=0
        temp_indice= -1
        temp_reg=0
        for k in range(np.shape(X)[1]):
            if not(k in deja_utilise):
                a=OLS_per_covariates(X,Y_temp,k)
                if np.abs(stat_of_no_effect(X,Y_temp,k,a))> temp_valeur :
                    temp_valeur=np.abs(stat_of_no_effect(X,Y_temp,k,a))
                    temp_indice=k
                    temp_reg=a
        res.append((temp_indice,p_value_bis(X,Y_temp,temp_indice,temp_reg)))
        deja_utilise.append(temp_indice)
        Y_temp = Y_temp - temp_reg.coef_[0]*X[:][temp_indice]
    return res 
        
            
        
        
        

    
```

```python
L=forward_variable_selection(X_train,Y_train)
#stat_of_no_effect(X_train,Y_train,1,OLS_per_covariates(X_train,Y_train,1))
L
```

### 3) 

```python
def selection_p_value(L):
    res=[]
    i=0
    while L[i][1]<0.05:
        res.append(L[i][0])
        i+=1
    return res
def creation_X(X,L):
    n=np.shape(X)[0]
    p=len(L)
    A=np.ones((p,n))
    for i in range(len(L)):
        A[:][i]=np.reshape(selection_colonne(X,L[i]),(1,-1))
    return A.T
reg=LinearRegression(fit_intercept=False).fit(creation_X(X_train,selection_p_value(L)),Y_train)
res=reg.score(creation_X(X_test,selection_p_value(L)),Y_test)
new_row = pd.DataFrame({'Methods':['fwd_selection'],'R2': [res]})
df_coef = df_coef.append(new_row, ignore_index=True)
df_coef
```

## 4) 

```python
sfs = SequentialFeatureSelector(LinearRegression(fit_intercept=False), n_features_to_select=22)
sfs.fit(X_train,Y_train)
H=sfs.get_support()
def transform(H):
    res=[]
    for i in range(len(H)):
        if H[i]:
            res.append(i)
    return res 
reg=LinearRegression(fit_intercept=False).fit(creation_X(X_train,transform(H)),Y_train)
```

```python
reg.score(creation_X(X_test,transform(H)),Y_test)
res=reg.score(creation_X(X_test,transform(H)),Y_test)
new_row = pd.DataFrame({'Methods':['fwd_selection_seq'],'R2': [res]})
df_coef = df_coef.append(new_row, ignore_index=True)
df_coef
```

The two algorithmes do not return the same variables because they do not select the variables the same way. The forward variable selection is based on test hypothesis whereas sequential feature selection is based on the cross-validation score of an estimator.

## Ridge 


### 5)

### a)

```python
def ridge(X,Y,alpha):
    n=np.shape(X)[0]
    p=np.shape(X)[1]
    gram_design= X.T@X
    return np.linalg.inv(gram_design+n*alpha*np.eye(p))@(X.T)@Y
x = np.logspace(-9, 2, 100)
p=np.shape(X_train)[1]
A=np.ones((p,len(x)))

for i in range(len(x)):
    L=ridge(X_train,Y_train,x[i])
    for j in range(p):
        A[j][i]=L[j]


# création du graphique
for i in range(p) :
    plt.plot(x,A[i])
    

plt.xscale('log')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphique du ridge en fonction de alpha')
plt.show()
    
```

### b) 

```python
def MSE_prediction_Ridge(X,Y,alpha):
    n=np.shape(X)[0]
    L=Y-X@(ridge(X_train,Y_train,alpha).T)
    return (np.linalg.norm(L)**2)/n
def min_with_index(arr):
    min_val = min(arr)
    min_idx = arr.index(min_val)
    return min_val, min_idx

```

```python
x = np.logspace(-9, 2, 100)
Y=[]
Y_bis=[]
for i in range(len(x)):
    Y.append(MSE_prediction_Ridge(X_train,Y_train,x[i]))
    Y_bis.append(MSE_prediction_Ridge(X_test,Y_test,x[i]))
Y_min,X_min=min_with_index(Y)
Y_min_bis,X_min_bis=min_with_index(Y_bis)
plt.plot(x[X_min], Y_min, 'ro', markersize=10)
plt.plot(x[X_min_bis], Y_min_bis, 'bo', markersize=10)
plt.plot(x,Y,color='r',label='train')
plt.plot(x,Y_bis,color='b',label='test')
plt.xscale('log')
plt.xlabel('x')
plt.legend(loc='best')
plt.ylabel('y')
plt.title('Graphique MSE en fonction de alpha')
plt.show()
```

The choice of alpha seems strange maybe it is more accurate to select from a K cross validation over the training set. Here it is just the smaller alpha which minimize the MSE

```python
n=np.shape(X_test)[0]
R2=1-n*MSE_prediction_Ridge(X_test,Y_test,10**(-9))/(np.linalg.norm(Y_test-np.mean(Y_test))**2 )
#R2=1-n*MSE_prediction_Ridge(X_test,Y_test,0.0016681005372000625)/(np.linalg.norm(Y_test-np.mean(Y_test))**2 )
new_row = pd.DataFrame({'Methods':['Ridge'],'R2': [R2]})
df_coef = df_coef.append(new_row, ignore_index=True)
df_coef
```

## b) 

```python
def nbr_differents_zeros(L):
    res=0
    for  i in L : 
        if i!=0:
            res+=1
    return res
x = np.logspace(-5, -2, 100)
Y=[]
for i in x: 
    clf = linear_model.Lasso(fit_intercept=False,alpha=i,max_iter=10**4,tol=0.05) 
    clf.fit(X_train,Y_train)
    Y.append(nbr_differents_zeros(clf.coef_))
plt.plot(x,Y)
    

plt.xscale('log')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphique nombre de coeff non nul en fonction de alpha pour le train')
plt.show()
```

```python
def MSE_prediction_Lasso(X,Y,rho):
    n=np.shape(X)[0]
    clf = linear_model.Lasso(fit_intercept=False,alpha=rho,max_iter=10**4,tol=0.05) 
    clf.fit(X_train,Y_train)
    L=Y-clf.predict(X)
    return (np.linalg.norm(L)**2)/n
x = np.logspace(-5,-2, 100)
Y=[]
Y_bis=[]
for i in range(len(x)):
    Y.append(MSE_prediction_Lasso(X_train,Y_train,x[i]))
    Y_bis.append(MSE_prediction_Lasso(X_test,Y_test,x[i]))
Y_min,X_min=min_with_index(Y)
Y_min_bis,X_min_bis=min_with_index(Y_bis)
plt.plot(x[X_min], Y_min, 'ro', markersize=10)
plt.plot(x[X_min_bis], Y_min_bis, 'bo', markersize=10)
plt.plot(x,Y,color='r',label="train")
plt.plot(x,Y_bis,color='b',label='test')
plt.xscale('log')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.title('Graphique MSE en fonction de alpha pour le test en bleu et le train en rouge')
plt.show()
```

```python
n=np.shape(X_test)[0]
R2=1-n*MSE_prediction_Lasso(X_test,Y_test,x[X_min_bis])/(np.linalg.norm(Y_test-np.mean(Y_test))**2 )
new_row = pd.DataFrame({'Methods':['Lasso'],'R2': [R2]})
df_coef = df_coef.append(new_row, ignore_index=True)
df_coef
```

```python
def K_cross(K,liste_alpha,liste_l1,X,Y):
    l=len(liste_alpha)
    h=len(liste_l1)
    n=np.shape(X)[0]
    taille_validation=int(n/K)
    MSE_fold=np.ones((K,l*h))
    for i in range(K):
        for k in range(l):
            for j in range(h):
                #if liste_l1[j]!=0 or liste_alpha[i]!=0:#otherwise it's classic OLS
                regr = ElasticNet(alpha=liste_l1[j]+liste_alpha[k], l1_ratio=liste_l1[j]/(liste_l1[j]+liste_alpha[k]), fit_intercept=False,max_iter=10**4,tol=0.05)
                regr.fit(np.concatenate((X[:i*taille_validation], X[(i+1)*taille_validation:]), axis=0),np.concatenate((Y[:i*taille_validation], Y[(i+1)*taille_validation:]), axis=0))
                MSE_fold[i][k*h+j]=(np.linalg.norm(Y[i*taille_validation:(i+1)*taille_validation]-regr.predict(X[i*taille_validation:(i+1)*taille_validation]))**2)/taille_validation
    min_temp=10**5
    temp_indice=0
    for i in range(np.shape(MSE_fold)[1]):
        m=np.mean(MSE_fold[:,i])
        if min_temp>m:
            min_temp=m
            temp_indice=i
    a=temp_indice//h
    b=temp_indice%h
    return liste_alpha[a],liste_l1[b]
        
        
liste_alpha= np.logspace(-10,3, 100)
liste_l1=[0,0.1,0.5, 0.7, 0.9, 0.95, 0.99]

res=K_cross(10,liste_alpha,liste_l1,np.array(X_train),np.array(Y_train))
res
        
        
        
        
        
        
        
    
    
```
